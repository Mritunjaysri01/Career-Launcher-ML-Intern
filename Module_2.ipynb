```
ASHOKA_dt=pd.read_csv('ASHOKA.csv')
BAJAJELEC_dt=pd.read_csv('BAJAJELEC.csv')
BOMDYEING_dt=pd.read_csv('BOMDYEING.csv')
CENTURYPLY_dt=pd.read_csv('CENTURYPLY.csv')
FORTIS_dt=pd.read_csv('FORTIS.csv')
```

```
ASHOKA_dt=ASHOKA_dt[ASHOKA_dt['Series']=="EQ"]
BAJAJELEC_dt=BAJAJELEC_dt[BAJAJELEC_dt['Series']=="EQ"]
BOMDYEING_dt=BOMDYEING_dt[BOMDYEING_dt['Series']=="EQ"]
CENTURYPLY_dt=CENTURYPLY_dt[CENTURYPLY_dt['Series']=="EQ"]
FORTIS_dt=FORTIS_dt[FORTIS_dt['Series']=="EQ"]
```

```
ASHOKA_dt=ASHOKA_dt[['Close Price']]
ASHOKA_dt.columns=['ASHOKA']
BAJAJELEC_dt=BAJAJELEC_dt[['Close Price']]
BAJAJELEC_dt.columns=['BAJAJELEC']
BOMDYEING_dt=BOMDYEING_dt[['Close Price']]
BOMDYEING_dt.columns=['BOMDYEING']
CENTURYPLY_dt=CENTURYPLY_dt[['Close Price']]
CENTURYPLY_dt.columns=['CENTURYPLY']
FORTIS_dt=FORTIS_dt[['Close Price']]
FORTIS_dt.columns=['FORTIS']
```

```
Data=pd.concat([ASHOKA_dt,BAJAJELEC_dt,BOMDYEING_dt,CENTURYPLY_dt,FORTIS_dt],axis=1)
Data
```

```
Data_pc=Data.pct_change()*100
Data_pc.dropna(inplace=True, axis=0)
```

```
import seaborn as sns
sns.set_style('whitegrid')
sns.set(font_scale=1.25)
sns.pairplot(Data_pc)
```

```
FORTIS_roll=Data_pc['FORTIS'].rolling(7).std()*np.sqrt(7)
FORTIS_roll.plot(figsize=(15,8))
```

```
Nifty=pd.read_csv(r"Nifty50.csv")
```

```
N_data=Nifty[['Close']]
N_data.columns=['Nifty']
N_data.head()
```

```
new_data=pd.concat([N_data,BAJAJELEC_dt,FORTIS_dt],axis=1)
new_data.pct_change()*100
```

```
new_data.dropna(inplace=True, axis=0)
```

```
new_data.head()
```

```
Volatility=new_data.rolling(7).std()*np.sqrt(7)
Volatility.plot(figsize=(15,8))
```

```
short_window = 21
long_window = 34

signals = pd.DataFrame(index=new_data.index)
signals['signal'] = 0.0

signals['short_mavg'] = new_data["FORTIS"].rolling(window=short_window, min_periods=1,center=False).mean()


signals['long_mavg'] = new_data['FORTIS'].rolling(window=long_window,min_periods=1, center=False).mean()

signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] > signals['long_mavg'][short_window:], 1.0,0.0)

signals['positions'] = signals['signal'].diff()
signals
```

```

fig = plt.figure(figsize=(20,15))

ax1 = fig.add_subplot(111, ylabel='Price')

new_data['FORTIS'].plot(ax=ax1, color='black', lw=2.)

signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

ax1.plot(signals.loc[signals.positions == 1.0].index, signals.short_mavg[signals.positions == 1.0], '^' , markersize=20,color='g')

ax1.plot(signals.loc[signals.positions == -1.0].index, signals.short_mavg[signals.positions == -1.0], 'v' , markersize=20,color='r')

plt.show()
```

```

symbol = 'BAJAJELEC'

df = pd.read_csv('BAJAJELEC.csv'.format(symbol), index_col='Date',
                 parse_dates=True, usecols=['Date', 'Close Price'],
                 na_values='nan')

df = df.rename(columns={'Close Price': symbol})
df.dropna(inplace=True)

sma = df.rolling(window=14).mean()

rstd = df.rolling(window=14).std()

upper_band = sma + 2 * rstd
upper_band = upper_band.rename(columns={symbol: 'upper'})
lower_band = sma - 2 * rstd
lower_band = lower_band.rename(columns={symbol: 'lower'})
df = df.join(upper_band).join(lower_band)
ax = df.plot(title='{} Price and BB'.format(symbol))
ax.fill_between(df.index, lower_band['lower'], upper_band['upper'], color='#ADCCFF')
ax.set_xlabel('Date')
ax.set_ylabel('SMA and BB')
ax.grid()
plt.show()
```
